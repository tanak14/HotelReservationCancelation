```python
hb_cancel=pd.read_csv('hotel_bookings 2.csv')
```

```python
hb_cancel['reservation_status_date']=pd.to_datetime(hb_cancel['reservation_status_date'],format="%d/%m/%Y")
```

```python
hb_cancel.drop(['company','agent'], axis=1, inplace=True)
hb_cancel.dropna(inplace=True)  #it remove all null values from tha data
```

```python
hb_cancel.describe()   # it describes about numerical data
```

```python
hb_cancel=hb_cancel[hb_cancel['adr']<5000]
```

# Data Analysis and Visualizations

```python
# percentage of canceled bookings
can_per=hb_cancel['is_canceled'].value_counts(normalize = True)   # normalize in value count shows the percentage
```

```python
can_per
```

```python
can_per=hb_cancel['is_canceled'].value_counts(normalize = True).plot(kind='bar')
plt.title('Reservation status count')
plt.bar(['Not Canceled','Canceled'],hb_cancel['is_canceled'].value_counts(), edgecolor = 'k', width=0.7)
plt.xticks(['Not Canceled', 'Canceled'], ['Not Canceled', 'Canceled'], rotation=0)
plt.show()
```

```python
plt.figure(figsize = (8,4))
fig1=sns.countplot(x='hotel', hue='is_canceled', data=hb_cancel, palette='Blues')
legend_labels,_=fig1.get_legend_handles_labels()
fig1.legend(legend_labels, ['Not Canceled', 'Canceled'], bbox_to_anchor=(1, 1))
plt.title('Resevation status in all the hotels',size = 20)
plt.xlabel('hotel')
plt.ylabel('number of resevations')
plt.show()
```

```python
resort_hotel=hb_cancel[hb_cancel['hotel']=='Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize = True)
```

```python
city_hotel=hb_cancel[hb_cancel['hotel']=='City Hotel']
city_hotel['is_canceled'].value_counts(normalize = True)
```

```python
resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel = city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize = (20,6))
plt.title('Average Daily Rate In City And Resort Hotel', fontsize = 30)
plt.plot(resort_hotel.index, resort_hotel['adr'], label='Resort Hotel')
plt.plot(city_hotel.index, city_hotel['adr'], label='City Hotel')
plt.legend(fontsize=20)
plt.show()
```

```python
hb_cancel['month']=hb_cancel['reservation_status_date'].dt.month
plt.figure(figsize = (16,8))
fig1=sns.countplot(x='month', hue='is_canceled', data=hb_cancel, palette='Blues')
legend_labels,_=fig1.get_legend_handles_labels()
fig1.legend(legend_labels, ['Not Canceled', 'Canceled'], bbox_to_anchor=(1, 1))
plt.title('Resevation status in all the hotels',size = 20)
plt.xlabel('month')
plt.ylabel('number of resevations')
plt.show()
```

```python
plt.figure(figsize=(14,6))
plt.title('ADR per month', fontsize = 30)
sns.barplot(x='month', y='adr', data=hb_cancel[hb_cancel['is_canceled'] == 1].groupby('month')[['adr']].sum().reset_index())
plt.show()
```

```python
can_data=hb_cancel[hb_cancel['is_canceled']==1]
top_10_country = can_data['country'].value_counts()[:10]  # this will show the top 10 countries
plt.figure(figsize=(6,6))
plt.title('Top 10 countries with reservation canceled')
plt.pie(top_10_country,autopct='%.2f',labels =top_10_country.index)
plt.show()
```

```python
can_adr = can_data.groupby('reservation_status_date')[['adr']].mean()
can_adr.reset_index(inplace = True)
can_adr.sort_values('reservation_status_date', inplace = True)

noncan_data=hb_cancel[hb_cancel['is_canceled']==0]
noncan_adr =noncan_data.groupby('reservation_status_date')[['adr']].mean()
noncan_adr.reset_index(inplace = True)
noncan_adr.sort_values('reservation_status_date', inplace = True)

plt.figure(figsize = (20,6))
plt.title('Average Daily Rate', fontsize = 30)
plt.plot(noncan_adr['reservation_status_date'], noncan_adr['adr'], label='Not Cancelled')
plt.plot(can_adr['reservation_status_date'], can_adr['adr'], label='Cancelled')
plt.legend(fontsize=20)
plt.show()
```

```python
can_adr=can_adr[(can_adr['reservation_status_date']>'2016') & (can_adr['reservation_status_date']<'2017-09')]
noncan_adr=noncan_adr[(noncan_adr['reservation_status_date']>'2016') & (noncan_adr['reservation_status_date']<'2017-09')]
```

```python
plt.figure(figsize = (20,6))
plt.title('Average Daily Rate', fontsize = 30)
plt.plot(noncan_adr['reservation_status_date'], noncan_adr['adr'], label='Not Cancelled')
plt.plot(can_adr['reservation_status_date'], can_adr['adr'], label='Cancelled')
plt.legend(fontsize=20)
plt.show()
```
